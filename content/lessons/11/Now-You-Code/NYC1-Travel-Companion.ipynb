## Step 1: Problem Analysis `get_exchange_rates`

Inputs: None

Outputs: a dictionary of current exchange rates

Algorithm (Steps in Program): 

1. Define a function to convert exchange rates
2. Assign Variable to URL
3. Assign Variable to get response
4. Get response
5. Convert to json
 


````python
# Step 2: write code here for the user-defined function get_exchange_rates
import requests
import json

def get_exchange_rates():
    url = 'http://data.fixer.io/api/latest?access_key=35b855dbc27ee9068965dd70110063e7'
    response = requests.get(url)
    data = response.json()
    return data
    
    
````

## Step 4: Problem Analysis for entire program

Inputs: Float number of desired currency 

Outputs: Sentence with numbers

Algorithm (Steps in Program):

1. Ask user to input currency
2. Set a variable to call function
3. Call the function by using the variable
4. Use input variable to write formula
5. Print results
6. Test for exceptions


````python
url = 'http://data.fixer.io/api/latest?access_key=35b855dbc27ee9068965dd70110063e7' 
try:
    response = requests.get(url) 
    response.raise_for_status()   
    data = response.json()       

except requests.exceptions.RequestException as e: 
    print("ERROR: Cannot connect to ", url)
    print("Please check your connection and restart the program")
    
try:
    currency = float(input("Enter currency amount in Euros: "))
    data = get_exchange_rates()
    money1 = data['rates']['GBP']
    money2 = data['rates']['USD']
    GBP = currency * money1
    USD = currency * money2

    print('%.2f in EUR is' % currency, '%.2f in GBP' % GBP)
    print('%.2f in EUR is' % currency, '%.2f in USD' % USD)
except ValueError:
    print("This is not a currency amount")
````
